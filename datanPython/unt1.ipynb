## Типичные наблюдения

```python
INS.describe(include='all') 
```

<BIG> Типичное наблюдение возраста (ср арифметическое) удачно, так как сравнимо с медианой. Однако, среднее арифметическое сильно зависит от выбросов => Выбор типичного наблюдения зависит от задачи. Это заметно в отличии среднего и медианы в значениях трат

<BIG> Можно отбросить некоторое количество максимальных и минимальных значений, чтобы избавится от выбросов (усеченное среднее)

<BIG> Также можно использовать разность квантилей Q3-Q1, стандартное отклонение (корень из дисперсии), дисперсию - описывают разброс

## [Диаграммы рассеяния](https://ru.wikipedia.org/wiki/%D0%94%D0%B8%D0%B0%D0%B3%D1%80%D0%B0%D0%BC%D0%BC%D0%B0_%D1%80%D0%B0%D1%81%D1%81%D0%B5%D1%8F%D0%BD%D0%B8%D1%8F)

```python
INS.groupby('sex')['bmi'].plot.hist(alpha=0.6)
plt.legend();
```

```python
from pandas.plotting import scatter_matrix
colors = {'northeast': 'green', 'northwest': 'green', 'southeast': 'red', 'southwest': 'red'}
scatter_matrix(INS, figsize=(10,10), diagonal='kde', c=INS['region'].replace(colors), alpha = 0.6);
```

<BIG> Определим наиболее информативные переменные (траты/возраст и возраст/имт наиболее наглядны)

```python
INS.plot.scatter(x='bmi',y='age',c=INS['region'].replace(colors), alpha = 0.6);
```

<BIG> На севере люди имеют меньший имт. можно рассмотреть границы переходов

```python
INS.plot.scatter(x='age',y='charges',c=INS['region'].replace(colors), alpha = 0.6);
```

<BIG>Траты на лечение сильно зависят от возраста, но слабо от региона
